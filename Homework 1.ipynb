## 3a

```python
import numpy as np
def flip(p):
    if p < 0 or p > 1:
        raise ValueError("p not between 0 and 1")
    return np.random.random() <= p

print(flip(0.26))
```

## 3b

```python
def flipn(p, n):
    return np.sum(np.random.random(n) > p)
n_tails = flipn(0.67, 1000)
print(n_tails)
```
